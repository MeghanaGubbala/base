```python
a=[230,560,780,127,128]
b=[200,160,270,127,400]
years=[1,2,3,4]
profit_a=[(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b=[(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth='3',label='CompanyA',marker='*',ms='15',mec='m')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label='CompanyB',marker='H')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("/content/Temperature.txt",sep=" ")
print(df)
print("\n The average temperature in 50 days is:")
avg=df['Temperature'].mean()
print(avg)
print("\n The lowest temperature recorded is:")
print(df['Temperature'].min())
print("\n The highest temperature recorded is:")
print(df['Temperature'].max())
day=0
for i in df['Temperature']:
  if (i>avg):
    day=day+1
print("\n Number of days where the temperature exceeded treshold is:")
plt.plot(df['Days'],df['Temperature'],color='black')
plt.title("Tempetature for 50 days")
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("iris")
print(tips)
sns.scatterplot(x="sepal_length",y="sepal_width",data=tips)
plt.title("Scatter plot of sepal_length vs.sepal_width")
plt.xlabel("sepal_length($)")
plt.ylabel("sepal_width($)")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
sns.violinplot(x="day",y="total_bill",data=tips)
plt.title("Distribution of Total bill by day")
plt.xlabel("Day of the week")
plt.ylabel("Total bill($)")
plt.show()
```
